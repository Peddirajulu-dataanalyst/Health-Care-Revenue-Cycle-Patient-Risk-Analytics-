```python
import pandas as pd 
import numpy as np
from faker import Faker
import random

fake = Faker()

#Number of patients

num_patients = 50000

insurance_list = [
    "Aetna",
    "Blue Cross",
    "Cigna",
    "United Healthcare",
    "Medicure"
]

chronic_conditions = [
    "Diabetes",
    "Hypertension",
    "Heart Disease",
    "Asthma",
    "None"
]

patients = []

for i in range(1, num_patients + 1):

    patient = {
        "patient_id": i,
        "patient_name": fake.name(),
        "age": random.randint(18,90),
        "gender": random.choice(["Male", "Female"]),
        "city": fake.city(),
        "insurance_provider": random.choice(insurance_list),
        "chronic_condition": random.choice(chronic_conditions),
        "registration_date": fake.date_between(
            start_date= '-3y',
            end_date = 'today'
        )
    }

    patients.append(patient)

#convert to dataframe

patients_df = pd.DataFrame(patients)

#save csv

patients_df.to_csv(r"D:\python project files\Health Care Revenue Cycle & Patient Risk Analytics\data\raw\patients.csv", index = False)

print("patients.csv created sucessfully")
print(patients_df.head())
    
```

```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# Load patients so visits can link correctly
patients_df = pd.read_csv(
    r"D:\python project files\Health Care Revenue Cycle & Patient Risk Analytics\data\raw\patients.csv"
)

# Number of visits
num_visits = 200000

# Mapping chronic condition → department + diagnosis
condition_map = {
    "Diabetes": {"department": "General Medicine", "diagnosis": "Diabetes"},
    "Hypertension": {"department": "General Medicine", "diagnosis": "Hypertension"},
    "Heart Disease": {"department": "Cardiology", "diagnosis": "Chest Pain"},
    "Asthma": {"department": "Pulmonology", "diagnosis": "Asthma"},  # you can add Pulmonology
    "None": {"department": "General Medicine", "diagnosis": "Fever"}  # fallback
}

payment_status_list = ["Paid", "Pending", "Denied"]

visits = []

for i in range(1, num_visits + 1):
    # Pick a random patient record
    patient = random.choice(patients_df.to_dict('records'))
    chronic_condition = patient['chronic_condition']

    # 70% chance to follow mapping, 30% random for variety
    if chronic_condition in condition_map and random.random() < 0.7:
        dept = condition_map[chronic_condition]['department']
        diag = condition_map[chronic_condition]['diagnosis']
    else:
        # fallback random choice
        dept = random.choice([
            "Cardiology", "Orthopedics", "Neurology", "Emergency",
            "Pediatrics", "Oncology", "General Medicine"
        ])
        diag = random.choice([
            "Diabetes", "Hypertension", "Fracture", "Chest Pain",
            "Migraine", "Asthma", "Fever", "Cancer"
        ])

    visit = {
        "visit_id": i,
        "patient_id": patient['patient_id'],
        "visit_date": fake.date_between(start_date='-2y', end_date='today'),
        "department": dept,
        "doctor_id": random.randint(1001, 1100),
        "diagnosis": diag,
        "treatment_cost": round(random.uniform(100, 15000), 2),
        "payment_status": random.choice(payment_status_list)
    }

    visits.append(visit)

# Create dataframe
visits_df = pd.DataFrame(visits)

# Save csv
visits_df.to_csv(
    r"D:\python project files\Health Care Revenue Cycle & Patient Risk Analytics\data\raw\visits.csv",
    index=False
)

print("visits.csv created successfully")
print(visits_df.head())
```

```python
import pandas as pd
import random

num_claims = 120000

claim_status_list = ["Approved", "Pending", "Denied"]

denial_reasons = [
    "Incomplete Documents",
    "Policy Expired",
    "Coverage Issue",
    "Duplicate Claim",
    "Preauthorization Missing"
]

claims = []

for i in range(1, num_claims + 1):
    patient_id = random.randint(1, 50000)
    claim_amount = round(random.uniform(500, 20000), 2)
    claim_status = random.choice(claim_status_list)

    # Approved amount logic — INSIDE the loop
    if claim_status == "Approved":
        approved_amount = round(claim_amount * random.uniform(0.7, 1.0), 2)
        denial_reason = "None"
    elif claim_status == "Pending":
        approved_amount = 0
        denial_reason = "None"
    else:  # Denied
        approved_amount = 0
        denial_reason = random.choice(denial_reasons)

    claim = {
        "claim_id": i,
        "patient_id": patient_id,
        "claim_amount": claim_amount,
        "approved_amount": approved_amount,
        "claim_status": claim_status,
        "denial_reason": denial_reason,
        "days_to_settlement": random.randint(1, 90)
    }
    claims.append(claim)

claims_df = pd.DataFrame(claims)
claims_df.to_csv(r"D:\python project files\Health Care Revenue Cycle & Patient Risk Analytics\data\raw\claims.csv", index=False)
print("claims.csv created successfully")
print(claims_df.head())
```

```python
import pandas as pd
import random
from faker import Faker

fake = Faker()
num_followups = 100000

followups = []

for i in range(1, num_followups + 1):
    patient_id = random.randint(1, 50000)
    missed_followup = random.choice(["Yes", "No"])

    # Readmission logic — INSIDE the loop
    if missed_followup == "Yes":
        readmission_flag = random.choices(["Yes", "No"], weights=[70, 30])[0]
    else:
        readmission_flag = random.choices(["Yes", "No"], weights=[20, 80])[0]

    record = {
        "followup_id": i,  # Added an ID for clarity
        "patient_id": patient_id,
        "followup_date": fake.date_between(start_date='-2y', end_date='today'),
        "missed_followup": missed_followup,
        "readmission_flag": readmission_flag
    }
    followups.append(record)

followups_df = pd.DataFrame(followups)
followups_df.to_csv(r"D:\python project files\Health Care Revenue Cycle & Patient Risk Analytics\data\raw\followups.csv", index=False)
print("followups.csv created successfully")
print(followups_df.head())
```
